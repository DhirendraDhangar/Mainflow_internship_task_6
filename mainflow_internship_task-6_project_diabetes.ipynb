```python
#ploting the bar chart for the first question
df.Diagnosis.value_counts().plot(kind='bar',color=["orchid","salmon"])
plt.title("DIABETIES DIAGNOSIS")
plt.xlabel("1 =DIABETIES DIAGNOSIS, 0 = NO DIABETIES DIAGNOSIS")
plt.ylabel("Amount");
```
